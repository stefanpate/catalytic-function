```python
import pandas as pd
import matplotlib.pyplot as plt
from src.filepaths import filepaths
```

```python
df = pd.read_csv(
    filepath_or_buffer=filepaths.artifacts / "241212_negative_sampling_mlflow_pull.csv",
    sep=','
)
df.head()
```

```python
# Pivot

ptable = pd.pivot_table(
    data=df,
    index='data/neg_multiple',
    columns='training/pos_multiplier',
    values=['val/binary_precision', 'val/f1', 'val/binary_recall', 'val/prc', 'val/roc'],
    aggfunc='mean',
    sort=True
)
ptable
```

```python
# ptable = ptable.swaplevel(axis=1)
# ptable
```

```python
ptable.columns
```

```python
# Sort

col_srt_key = {
    'val/binary_precision': 0,
    'val/f1': 2,
    'val/binary_recall': 1,
    'val/prc': 3,
    'val/roc': 4
}

ptable = ptable.sort_index(axis=1, level=0, sort_remaining=False).sort_index(axis=1, level=1, key=lambda x : x.map(col_srt_key), sort_remaining=False)
ptable
```

```python
# Rename

col_rename = {
    'val/binary_precision': 'Precision',
    'val/f1': 'F1',
    'val/binary_recall': 'Recall',
    'val/prc': 'AUPRC',
    'val/roc': 'AUROC'
}
ptable.rename(
    columns=col_rename,
    level=0,
    inplace=True
)

ptable.index.name = "Negative multiple"
ptable.columns.names = [None, "Positive multiplier"]

ptable
```

```python
# Style

# idx = pd.IndexSlice # For greater flexibility selecting across 2 dims

s = ptable.style

s.format(precision=3)

metrics = set(ptable.columns.get_level_values(0))

for m in metrics:
    if m == 'Precision':
        continue

    s.set_table_styles(
        {
            (m, 1): [
                {'selector': 'th', 'props': 'border-left: 1px solid black'},
                {'selector': 'td', 'props': 'border-left: 1px solid #000066'}
            ]
        }, 
        overwrite=False,
        axis=0
    )


for m in metrics:
    s.highlight_max(axis=None, subset=[m], props='font-weight:bold;background-color:lightgreen')
s
```
